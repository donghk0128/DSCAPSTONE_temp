```python
import pandas as pd

# CSV 불러오기 (두 번째 줄을 헤더로 사용)
df = pd.read_csv("data/위치관련/시내주요기관(경찰·소방관서).csv", encoding="utf-8", header=1)

# 필요한 컬럼만 선택
df = df[["자치구별(2)", "소방본부", "소방서", "119안전센터"]]

# 컬럼명 변경 (깔끔하게)
df.columns = ["자치구", "소방본부", "소방서", "119안전센터"]

# 불필요한 행 제거 (소계, 기타)
df = df[~df["자치구"].isin(["소계", "기타"])]

# 인덱스 재설정
df = df.reset_index(drop=True)

# 확인
print(df.head())
```

119+안전센터+1개소당+시민수.csv를 이거랑 결합

```python
# 2️⃣ 추가 데이터 불러오기 (헤더 2줄 처리)
df_extra = pd.read_csv("data/위치관련/119+안전센터+1개소당+시민수.csv", encoding="utf-8", header=1)

# 3️⃣ 필요한 컬럼만 선택
df_extra = df_extra[[
    "자치구별(2)",
    "인구 (명)",
    "119안전센터 1개소당 담당인구 (명)"
]]

# 4️⃣ 컬럼명 정리
df_extra.columns = [
    "자치구",
    "인구",
    "센터당 담당인구"
]

# 5️⃣ 불필요 행 제거 (소계, 기타)
df_extra = df_extra[~df_extra["자치구"].isin(["소계", "기타"])]

# 6️⃣ 인덱스 리셋
df_extra = df_extra.reset_index(drop=True)

# 7️⃣ 병합 (자치구 기준)
df_merged = pd.merge(df, df_extra, on="자치구", how="left")

# 8️⃣ 저장
df_merged.to_csv("data/위치관련/자치구별 소방기관 정보_preprocessed.csv", index=False, encoding="utf-8-sig")

# 확인
print(df_merged.head())
```
